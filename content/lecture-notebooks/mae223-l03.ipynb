```python
import sympy as sm
import sympy.physics.mechanics as me
```

```python
D = me.ReferenceFrame('D')
```

Set $\hat{\mathbf{d}}_3 = \hat{\mathbf{a}}_3$

```python
q1, q2, q3 = sm.symbols('q1, q2, q3', real=True)
```

```python
A = D.orientnew('A', 'Axis', (q3, -D.z))
```

```python
B = A.orientnew('B', 'Axis', (sm.pi/2-q2 , A.y))
```

```python
C = B.orientnew('C', 'Axis', (q1, B.x))
```

```python
R = sm.symbols('R')
```

```python
p = R*C.y
p
```

```python
p.express(B)
```

```python
p.express(A)
```

```python
p.express(D)
```

```python
p.free_symbols(D)
```

```python
p.diff(q1, A)
```

```python
p.diff(q2, B)
```

```python
p.express(D).simplify()
```

```python
p.diff(q1, A).simplify()
```

```python
w = R*B.x + R**2*C.y + R**3*A.x
w
```
